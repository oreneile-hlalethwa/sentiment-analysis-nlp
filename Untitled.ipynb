```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make charts look nice
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

# Load exported data
model_comparison = pd.read_csv('exports/model_comparison.csv')
class_distribution = pd.read_csv('exports/class_distribution.csv')
confusion_matrix_df = pd.read_csv('exports/confusion_matrix.csv')
top_words = pd.read_csv('exports/top_words_by_sentiment.csv')

print(" Data loaded successfully!")
```

```python
# Model Accuracy Comparison
plt.figure(figsize=(10, 6))
colors = ['#e74c3c', '#3498db', '#3498db', '#3498db', '#2ecc71']
bars = plt.barh(model_comparison['Model'], model_comparison['Test_Accuracy'], color=colors)
plt.xlabel('Test Accuracy (%)')
plt.title(' Model Accuracy Comparison', fontsize=16, fontweight='bold')
plt.xlim(0, 100)

# Add value labels on bars
for bar, val in zip(bars, model_comparison['Test_Accuracy']):
    plt.text(val + 1, bar.get_y() + bar.get_height()/2, f'{val}%', va='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Sentiment Distribution
plt.figure(figsize=(8, 8))
colors = ['#2ecc71', '#e74c3c', '#f39c12']
explode = (0.02, 0.02, 0.02)

plt.pie(class_distribution['Count'], 
        labels=class_distribution['Sentiment'],
        autopct='%1.1f%%',
        colors=colors,
        explode=explode,
        shadow=True,
        startangle=90)

plt.title(' Sentiment Distribution in Dataset', fontsize=16, fontweight='bold')
plt.show()
```

```python
# Confusion Matrix
cm_pivot = confusion_matrix_df.pivot(index='Actual', columns='Predicted', values='Count')

plt.figure(figsize=(8, 6))
sns.heatmap(cm_pivot, annot=True, fmt='d', cmap='Blues', cbar=True)
plt.title(' Confusion Matrix (TF-IDF Model)', fontsize=16, fontweight='bold')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.tight_layout()
plt.show()
```

```python
# Top Words for Each Sentiment
fig, axes = plt.subplots(1, 3, figsize=(15, 6))

sentiments = ['Positive', 'Negative', 'Neutral']
colors = ['#2ecc71', '#e74c3c', '#f39c12']

for ax, sentiment, color in zip(axes, sentiments, colors):
    data = top_words[top_words['Sentiment'] == sentiment].head(10)
    ax.barh(data['Word'], data['Weight'], color=color)
    ax.set_title(f'Top {sentiment} Words', fontweight='bold')
    ax.invert_yaxis()

plt.suptitle(' Most Important Words by Sentiment', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
